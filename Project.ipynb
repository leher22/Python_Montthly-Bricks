#### CHECKING FOR ANY STRING VALUES IN THE NUMERIC COLUMNS

```python
#Defining a function that takes 2 arguments
def check_for_string_values_in_numeric_columns(df, numeric_columns):
    # Create an empty dictionary to store information about string values found in numeric columns
    string_values_info = {}

    # Print a message to let us know that we're starting to check for string values
    print("\n=== Checking for String Values in Numeric Columns ===")
    
    # Looping through each numeric column (list of columns passed as 'numeric_columns')
    for col in numeric_columns:
        # Check each value in the column to see if it is NOT a number (i.e., it is a string or some other type)
        non_numeric_rows = df[~df[col].apply(lambda x: isinstance(x, (int, float)))]

        # If there are rows with non-numeric (string or other) values, we store details in the dictionary
        if not non_numeric_rows.empty:
            string_values_info[col] = {
                "Total String Values": len(non_numeric_rows),  # Total count of non-numeric values
                "Example Values": non_numeric_rows[col].head().tolist()  # Show a few example non-numeric values
            }

            # Print the details about the string values found
            print(f"\nColumn: {col}")
            print(f" - Total String Values Found: {len(non_numeric_rows)}")  # How many non-numeric values
            print(f" - Example String Values: {non_numeric_rows[col].head().tolist()}")  # Example values
        else:
            # If no string values are found, print a message saying so
            print(f"\nColumn: {col}")
            print(" - No string values found.")
    
    # Returns the dictionary containing details about any string values found in numeric columns
    return string_values_info


# Defining the list of columns that should contain numeric values
numeric_columns = [
    'Commons production', 'Facings production', 'Engineering production',
    'Total production', 'Commons deliveries', 'Facings deliveries',
    'Engineering deliveries', 'Total deliveries', 'Commons stocks',
    'Facings stocks', 'Engineering stocks', 'Total stocks'
]

# Calling the function to check for string values in these numeric columns
string_values_detected = check_for_string_values_in_numeric_columns(df, numeric_columns)

```

The `lambda` function checks each value in the specified column (df[col]) to see if it's numeric or NaN. The `~` negates the result, so that non_numeric_rows contains rows where the column values are not numeric.

As it is observed by the output there are no string values in the numeric columns

#### DETECTING OUTLIERS USING THE INTER QUARTILE RANGE (IQR) METHOD

In IQR method these are the following steps that should be followed:-

**Step 1:-**
Calculate the quartiles i.e. the first quartile **(Q1)** and third quartile **(Q3)**

**Step 2:-** Compute the IQR 

IQR = Q3-Q1

**Step 3**:- Determine the outlier threshold

For lower bound:- Q1-1.5 x IQR

For upper bound:- Q3 + 1.5 x IQR

**Step 4** :- Flagging Outliers

Any data point that falls below the lower bound or above the upper bound is considered an outlier

```python
#Defining a function that takes 1 argument that is being analysed 
def detect_outliers_iqr(df):
    outliers_info = {}  #empty dictionary that will be used to store information about the outliers for each column. 

    print("=== Outlier Detection Report ===")
    for col in numeric_columns:
        Q1 = df[col].quantile(0.25) #representing the lower quartile
        Q3 = df[col].quantile(0.75) #representing the upper quartile
        IQR = Q3 - Q1 #Computing the IQR
        lower_bound = Q1 - 1.5 * IQR  #any value below this threshold is considered an outlier
        upper_bound = Q3 + 1.5 * IQR  #any value above this threshold is considered an outlier

        # Identify outliers
        #filters df to find rows where the values in the column are either less than the lower bound or greater than the upper bound
        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

        # Storing outliers details
        outliers_info[col] = {
            'count': len(outliers),  #gives back the number of outliers detected in a specefic column
            'values': outliers[col].tolist()  #gives back the list of outlier values
        }

        # Checks if any outliers are present
        if len(outliers) > 0:  #if present then:
            print(f"\nColumn: {col}")  #prints the column name
            print(f" - Outliers Detected: {len(outliers)}") #the number of outliers
            print(f" - Outlier Values: {outliers[col].tolist()}")  #list of values of the outliers
        else:    #if no outliers are found
            print(f"\nColumn: {col}")  #prints column name
            print(" - No outliers detected.")  
#Retrieving the total number of outliers found
    total_outliers = sum([info['count'] for info in outliers_info.values()])
    print(f"\n=== Summary ===\nTotal Outliers Detected: {total_outliers}")

    return outliers_info  #returns the outlier dictionary, which will contain information about outliers

# Calling the function
outliers_detected = detect_outliers_iqr(df)
```

Now since there are outliers so we are checking if the distribution is normal or not. If the distribution is normal then we will replace the outliers with the **mean** or else we can use the **capping method**.

```python
#Checking for NORMAL DISTRIBUTION
#importing the required libraries
import seaborn as sns
import matplotlib.pyplot as plt

#defining a function that takes 1 argument 
def check_normality(df):

    for column in numeric_columns:
        print(f"\nAnalyzing column: {column}")  #looping through each column, and for each column prints a message to indicate which column is being analysed
        
        # Creating a histogram 
        plt.figure(figsize=(10, 5))  #creates a figure with 10 inches wide and 5 inches tall
        sns.histplot(df[column], bins=20, color='skyblue', edgecolor='black', alpha=0.8)
                                        
        # Add titles and labels for better readability
        plt.title(f'Distribution of {column}', fontsize=16, weight='bold', color='darkblue') #adding title 
        plt.xlabel(column, fontsize=12, weight='bold') #adding xlabel
        plt.ylabel('Frequency', fontsize=12, weight='bold') #adding ylabel
        
        # Add grid lines to make the chart more readable
        plt.grid(axis='y', linestyle='--', linewidth=0.7, alpha=0.6)
        
        # Adjust spacing for a cleaner look
        plt.tight_layout()
        
        # Show the plot
        plt.show()

# calling the function
check_normality(df)
```

*Since the dataset is not normally distributed so we will not replace the values by mean instead we will cap the outliers*


```python
#Defining a function
def cap_outliers_iqr(df):
    capped_data = []   #list that stores information about the outlier capping process 
    affected_rows = [] #list that will store the rows containing outliers

    
    for col in numeric_columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Identify rows that are outliers
        outlier_rows = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

        # Store the affected rows
        affected_rows.append(outlier_rows)  

        # Capping outliers
        df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)  


    # Combine the affected rows from all columns into one DataFrame
    affected_df = pd.concat(affected_rows).drop_duplicates()

    # Display capped data in a DataFrame
    capped_data_df = pd.DataFrame(capped_data)
    
    return df, affected_df

# Calling the functions
df_capped, affected_rows_df = cap_outliers_iqr(df)
```

`clip()` is used to cap the outliers. Any values below the lower bound will be set to the lower bound, and any values above the upper bound will be set to the upper bound. This ensures that no values fall outside the calculated range.

`drop_duplicates()` ensures that duplicate rows are removed, leaving only unique outlier rows.

DOING A FINAL CHECK IF ANY OUTLIERS ARE STILL LEFT OR NOT

```python
def detect_remaining_outliers(df):
    print("\n=== Checking for Remaining Outliers ===")
    remaining_outliers = []
    
    for col in numeric_columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Find any remaining outliers
        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        
        if len(outliers) > 0:
            remaining_outliers.append({
                "Column": col,
                "Outliers Count": len(outliers),
                "Outlier Values": outliers[col].tolist()
            })

    if remaining_outliers:
        print("\nRemaining Outliers Detected:")
        remaining_outliers_df = pd.DataFrame(remaining_outliers)
        print(remaining_outliers_df.to_string(index=False))
    else:
        print("\nNo remaining outliers detected in the dataset.")

# Check for remaining outliers
detect_remaining_outliers(df_capped)
```

##### CREATING A CORRELATION MATRIX

```python
#definig a function that takes 2 arguments
def plot_correlation_matrix(df, numeric_columns):
    # Calculate the correlation matrix
    correlation_matrix = df[numeric_columns].corr()

    # Plotting the heatmap for the correlation matrix
    plt.figure(figsize=(12, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
    
    # Adding title and labels
    plt.title("Correlation Matrix of Numeric Columns", fontsize=16, fontweight='bold')
    plt.xlabel("Columns", fontsize=14, fontweight='bold')
    plt.ylabel("Columns", fontsize=14, fontweight='bold')

    # Display the plot
    plt.tight_layout()
    plt.show()

# Call the function to plot the correlation matrix and generate the heatmap
plot_correlation_matrix(df, numeric_columns)
```

`corr()`, computes the pairwise correlation coefficients for the numeric columns. The result is a square matrix, where:- **Rows and Columns** are the numeric columns and values range from -1(as one variable decreases the pther also decreases) to 1(as one variable increases other also increases) with 0 indicating no relation.

`annot = True`, displays the correlation coefficient values in the heatmap. 

`cmap="coolwarm` uses the **coolwarm** colormap to represent values, with warm colors for positive correlations(red) and cool colors(blue) for negative correlations, white/neutral for no significant reations

`fmt=".2f`, formats the correlation values to two decimal places.

`linewidths=0.5`, adds thin lines between cells for better separation

#### Creating new columns like Cost of Goods Sold(COGS), Opening Stock and Closing Stock for getting the Stock Turnover Ratio(STR)

The **Stock Turnover Ratio** is a key performance metric that measures how efficiently a business uses ints inventory. This ratio is highly useful for operational analysis and decision making in supply chain. 

**COGS**:- It is the direct costs incurred to produce the goods sold.  

**StockTurnoverRatio=  COGS/Average Inventory**
 
**Average Inventory** :- The mean stock level during the period

**Average Inventory = (Opening Stock + Closing Stock)/2**

**Opening Stock** is the value of inventory that a business has at the beginning of an accounting period 

**Closing Stock** is the value of inventory remaining at the end of an accounting period after all sales, production, and inventory changes have been accounted for.

```python
# Clean the 'Month' column by stripping any extra spaces or unexpected characters
df['Month'] = df['Month'].str.strip()

# Ensure the data is sorted by Year and Month to maintain temporal order
df['Month Numeric'] = pd.to_datetime(df['Month'], format='%B', errors='coerce').dt.month
df = df.sort_values(by=['Year', 'Month Numeric']).reset_index(drop=True) #drop ensures that the old index is not added with the new sequential index

# Create Opening Stock
df['Opening Stock'] = df['Total stocks'].shift(1)  # Shift the Closing Stock to get Opening Stock
df.loc[0, 'Opening Stock'] = df.loc[0, 'Total stocks']  # Set the first row's Opening Stock equal to Closing Stock

# Closing Stock is already available as 'Total stocks'
df['Closing Stock'] = df['Total stocks']

# Calculate COGS (Cost of Goods Sold)
df['COGS'] = df['Opening Stock'] + df['Total production'] - df['Closing Stock']

# Display the first few rows of the updated dataset to verify the new columns
print("Dataset with Opening Stock, Closing Stock, and COGS:")
print(df[['Year', 'Month', 'Opening Stock', 'Closing Stock', 'COGS']].head())
```

`format='%B'` specifies that the input format is a full month name.

`errors='coerce'` ensures that if a value in the Month column is invalid, it is replaced with NaT (Not a Time).



```python
# Adding an Average Inventory column
df['Average Inventory'] = (df['Opening Stock'] + df['Closing Stock']) / 2

# Calculating Stock Turnover Ratio
df['Stock Turnover Ratio'] = df['COGS'] / df['Average Inventory']


# Display the first few rows to verify calculations
print("Dataset with Stock Turnover Ratio:")
print(df[['COGS', 'Opening Stock', 'Closing Stock', 'Average Inventory', 'Stock Turnover Ratio']].head())
```
